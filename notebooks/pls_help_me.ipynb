```
from smiles_transformer.tokenizer import *
from smiles_transformer.preprocessing.transform import *

tokenizer = IndividualTokenizer(vocab_file=r"C:\git\smiles-cgr-transformers\data\vocab\vocab_individual.txt")
```

```
import pandas as pd

df = pd.read_csv(r"C:\git\smiles-cgr-transformers\data\interim\e2sn2.csv")["AAM"].head(20)
df
```

```
cgr_tr = SMILEStoCGRTransform()
smi_cgr_tr = CGRtoStrTransform()
df = cgr_tr.transform(pd.DataFrame(df), in_column="AAM", out_column="CGR")
df = smi_cgr_tr.transform(pd.DataFrame(df), in_column="CGR", out_column="SMI_CGR")
import random
```
